Importing the required libraries

```python
import torch
from datasets import Dataset
from unsloth import FastLanguageModel
import json
from transformers import TrainingArguments
import time
import os
from google.colab import drive
import shutil
from datetime import datetime
import json
import re
import threading
```

Opening the dataset file. The dataset was scraped seperately and stored in a json file.

```python
with open('cleaned_training_data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

Creating the dataset from the list

```python
dataset = Dataset.from_list(data)
```

Pulling the model from huggingface under the name of unsloth and using the 4 bit bits&bytes quantized model.

```python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Meta-Llama-3.1-8B-bnb-4bit",
    max_seq_length=2048,
    dtype="bfloat16",
    load_in_4bit=True
)
```

In this project we are using parameter efficient fine tuning. Quantization is a type of peft training method. Here the configurations of peft are defined with no dropouts.

```python
peft_config = {
    "r": 16,
    "alpha": 16,
    "target_modules": ["all_linear"],
    "bias": False,
    "dropout": 0.0,
    "use_rslora": True
}
```

Loading the model using unsloth. Unsloth makes traing the model faster. It uses FastLanguageModel to load the model. Using QLora in this project. Qlora is a type of peft. Unlike traditional finetuning tasks, we only calculate the loss on the adapters of the quantized weights and then optimize them. This reducues the grpahical memory requirements significantly.

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                   "gate_proj", "up_proj", "down_proj"],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    use_rslora=True,
    loftq_config=None
)
```

First saving the model locally.

```python
LOCAL_SAVE_PATH = "/content/ros2_llama_finetuned"
```

Passing the training arguments to a variable which we will use further. Here the learning rate is 2e-4 and we are using the paged adamw 32 bit as the optimizer which is used mainly for 32bit models for optimization. Learning scheduler is cosine. Saving steps after evry 50 steps.

```python
training_args = TrainingArguments(
    output_dir=LOCAL_SAVE_PATH,
    num_train_epochs=6,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    learning_rate=2e-4,
    max_grad_norm=0.3,
    weight_decay=0.01,
    optim="paged_adamw_32bit",
    lr_scheduler_type="cosine",
    logging_steps=10,
    save_strategy="steps",
    save_steps=50,
    eval_strategy="no",
    fp16=True,
    report_to="tensorboard"
)
```

```python
!pip install trl
from trl import SFTTrainer
```

Using the SFT trainer module from unsloth and starting the training. As we can see the training loss drops from 1.6 to 0.3 in 6 epochs. 

```python
def formatting_func(examples):
    texts = []
    for i in range(len(examples['instruction'])):
        text = f"### Instruction: {examples['instruction'][i]}\n"
        text += f"### Response: {examples['output'][i]}"
        texts.append(text)
    return texts

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    tokenizer=tokenizer,
    max_seq_length=2048,
    args=training_args,
    formatting_func=formatting_func
)
```

```python
try:
    trainer.train()
except Exception as e:
    print(f"Training error: {str(e)}")
finally:
    # Save final model
    final_save_path = LOCAL_SAVE_PATH
    trainer.save_model(final_save_path)
    print(f"Final model saved to: {final_save_path}")
```
